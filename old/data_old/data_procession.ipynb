```python
import pandas as pd

# 读取数据
data_raw = pd.read_excel('data_without_dopants.xlsx')
data_raw.head()
```

```python
# 列清理
data = data_raw.drop(["Index", "Cathode", "Smicro", "Smeso", "Vmicro", "Vmeso", "Electrolyte", "Ref"], axis=1)
data
```

```python
# 数据分布统计
statistics = data.describe(include='all')
statistics
```

```python
# 填充缺失值
data["AML"] = data["AML"].fillna(data["AML"].median())

# 将Anion列的数据转换为数值
data["Anion"] = data["Anion"].map({"SO4": 1, "OTf": 0})

# 将CC列删除
data=data.drop(["CC"], axis=1)

data
```

```python
# 保存一份数据
data.to_csv('dataset.csv', index=False)
```
